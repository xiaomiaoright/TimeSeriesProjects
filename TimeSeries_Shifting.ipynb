# Shifting
Two methods: .shift() and .tshift()
* set the frequency of index: df.asfreq('D', method='pad')

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn
seaborn.set()
%matplotlib inline
```

## Load dataset

```python
from pandas_datareader import data

df = data.DataReader('GOOG', start='2004', end='2020', data_source='yahoo')
df.head()
```

```python
df.tail()
```

## Shift foreward backward
Series.shift(self, periods=1, freq=None, axis=0, fill_value=None)[source]

```python
df.shift(1).head()
```

```python
df.shift(-1).tail()
```

```python
df.shift(periods=1, freq='M').head()
```

```python
df.index
```

## tShift 

```python
# Need to set the index frequecy
df2 = df.asfreq('D', method='pad')
df2.index
```

```python
df2.head(20)
```

```python
df2.tshift(1).head()
```

```python
df.head()
```

```python
df2.tshift(-1).head()
```

```python
df2[:30]
```

```python
df2[:30]['Close'].plot(style = '-o', figsize=(12,6))
```

## Application: ROI

```python
ROI = 100 * (df2['Close'].tshift(-365) / df2['Close'] - 1)
ROI.plot()
plt.ylabel('% Return on Investment')
```
