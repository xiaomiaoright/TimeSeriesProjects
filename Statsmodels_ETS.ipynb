## Example of Airline passenger data

```python
df = pd.read_csv('Dataset\Airline_passengers.csv', index_col='Month', parse_dates=True)
df.head()
```

```python
df.info()
```

```python
# df.dropna(inplace=True) 
df.plot(figsize=(12,6))
```

## Statsmodel for seasonal decomposition

```python
from statsmodels.tsa.seasonal import seasonal_decompose
df_decomp = seasonal_decompose(df['Thousands of Passengers'], model='multiplicative')
```

```python

ax = df_decomp.observed.plot(figsize = (12,12), legend='Observed')
ax = df_decomp.trend.plot(legend = 'Trend')
ax.set(ylabel = 'Thousands of Passengers')
# ax.set_legend()
```

```python
df_decomp.seasonal.plot()
df_decomp.resid.plot()
```

```python
df_decomp.weights.plot()
```
