# Time Series Visualization and Customizing

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn
seaborn.set()
%matplotlib inline
```

```python
from pandas_datareader import data

df = data.DataReader('GOOG', start='2004', end='2020', data_source='yahoo')
df.head()
```

```python
ax = df[df.index.year==2004][['High','Low','Open','Close']].plot(figsize=(12,6))
ax.set( title='2004 google stock', xlabel='Date', ylabel='Price', alpha=0.5)
```

# Time Series Data Formating

## Add title and lables

```python
ax = df[df.index.year==2004][['High','Low','Open','Close']].plot(figsize=(12,6))
ax.set( title='2004 google stock', xlabel='Date', ylabel='Price', alpha=0.5)
ax.autoscale(axis='x',tight=True)
```

## Set XLimits Y limits
* by slicing data
* by xlim and ylim arguments

```python
True & False
```

```python
# by slicing data
# ax = df[(df.index.year==2004) & (df.index.month==12) ][['High','Low','Open','Close']].plot(figsize=(12,6))
ax = df['2004-12-01':'2004-12-31'][['High','Low','Open','Close']].plot(figsize=(12,6))
ax.set( title='2004 google stock', xlabel='Date', ylabel='Price', alpha=0.5)
ax.autoscale(axis='x',tight=True)
```

```python
# by xlims ylims arguments
ax = df['Close'].plot(figsize=(12,6),xlim=['2004-12-01','2004-12-31'], ylim=[80,100])
ax.set( title='2004 google stock', xlabel='Date', ylabel='Price', alpha=0.5)
# ax.autoscale(axis='y', tight=True)
```

## Color and Style


```python
ax = df['2004-12-01':'2004-12-31'][['High','Low','Open','Close']].plot(figsize=(12,6), style=[':', '--','-','-.'], color=['r','g','y','b'])
ax.set( title='2004 google stock', xlabel='Date', ylabel='Price', alpha=0.5)
ax.autoscale(axis='x',tight=True)
```

## X Ticks
Date format follows [strftime code](https://strftime.org/)

```python
from datetime import datetime
datetime(2001, 2, 3, 16, 5, 6).strftime("%A, %B %d, %Y  %I:%M:%S %p")
```

```python
from matplotlib import dates

ax = df['2004-12-01':'2004-12-31'][['High','Low','Open','Close']].plot(figsize=(12,6), style=[':', '--','-','-.'], color=['r','g','y','b'])
ax.set( title='2004 google stock', xlabel='\n Date 2014', ylabel='Price', alpha=0.5)
ax.autoscale(axis='x',tight=True)
ax.xaxis.set_major_locator(dates.WeekdayLocator(byweekday=0))
ax.xaxis.set_major_formatter(dates.DateFormatter("%a-%B-%d"))
```

## Major vs Minor Axis Values

```python
from matplotlib import dates

ax = df['2004-10-01':'2004-12-31'][['High','Low','Open','Close']].plot(figsize=(12,6), style=[':', '--','-','-.'], color=['r','g','y','b'])
ax.set( title='2004 google stock', xlabel='\n Year 2014', ylabel='Price', alpha=0.5)
ax.autoscale(axis='x',tight=True)
ax.xaxis.set_major_locator(dates.WeekdayLocator(byweekday=0))
ax.xaxis.set_major_formatter(dates.DateFormatter("%d"))

ax.xaxis.set_minor_locator(dates.MonthLocator())
ax.xaxis.set_minor_formatter(dates.DateFormatter('\n\n%B'))
```

## Grids

```python
from matplotlib import dates

ax = df['2004-10-01':'2004-12-31'][['High','Low','Open','Close']].plot(figsize=(12,6), style=[':', '--','-','-.'], color=['r','g','y','b'])
ax.set( title='2004 google stock', xlabel='\n Year 2014', ylabel='Price', alpha=0.5)
ax.autoscale(axis='x',tight=True)
ax.xaxis.set_major_locator(dates.WeekdayLocator(byweekday=0))
ax.xaxis.set_major_formatter(dates.DateFormatter("%d"))

ax.xaxis.set_minor_locator(dates.MonthLocator())
ax.xaxis.set_minor_formatter(dates.DateFormatter('\n\n%B'))

ax.yaxis.grid(False)
ax.xaxis.grid(False)
```
